```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# One-hot encode
df_encoded = pd.get_dummies(df, drop_first=True)

# Scale numerical features
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df_encoded)

# K-Means model
kmeans = KMeans(
    n_clusters=5,
    init='k-means++',
    max_iter=300,
    n_init=10,
    random_state=42
)
labels = kmeans.fit_predict(scaled_data)

# Evaluation
sil = silhouette_score(scaled_data, labels)
ch  = calinski_harabasz_score(scaled_data, labels)
db  = davies_bouldin_score(scaled_data, labels)

print("Silhouette Score:", sil)
print("Calinski-Harabasz:", ch)
print("Davies-Bouldin:", db)

# PCA Visualization
pca = PCA(n_components=2)
pca_values = pca.fit_transform(scaled_data)

plt.figure(figsize=(8,6))
plt.scatter(pca_values[:,0], pca_values[:,1], c=labels, s=50)
plt.title("K-Means Clusters (PCA Visualization)")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()
```

```python
from sklearn.decomposition import PCA
from sklearn.mixture import GaussianMixture
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score
import matplotlib.pyplot as plt

pca = PCA(n_components=2)
df_pca = pca.fit_transform(df_encoded)

print("Explained Variance Ratio PCA:", pca.explained_variance_ratio_)

gmm = GaussianMixture(
    n_components=5,
    covariance_type='full',
    max_iter=300,
    init_params='kmeans',
    random_state=42
)

gmm_labels = gmm.fit_predict(df_encoded)

gmm_sil = silhouette_score(df_encoded, gmm_labels)
gmm_ch = calinski_harabasz_score(df_encoded, gmm_labels)
gmm_db = davies_bouldin_score(df_encoded, gmm_labels)

print("Silhouette Score:", gmm_sil)
print("Calinski-Harabasz:", gmm_ch)
print("Davies-Bouldin:", gmm_db)

plt.figure(figsize=(8,6))
plt.scatter(df_pca[:,0], df_pca[:,1], c=gmm_labels, s=50)
plt.title("Gaussian Mixture Clustering (Visualized with PCA)")
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import pandas as pd

# Preprocessing
df_encoded = pd.get_dummies(df, drop_first=True)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_encoded)

# Agglomerative Clustering model (versi terbaru)
agg = AgglomerativeClustering(n_clusters=5, linkage='ward')  # 'affinity' tidak digunakan lagi
agg_labels = agg.fit_predict(X_scaled)

# Evaluation
sil = silhouette_score(X_scaled, agg_labels)
ch = calinski_harabasz_score(X_scaled, agg_labels)
db = davies_bouldin_score(X_scaled, agg_labels)

print("Silhouette Score:", sil)
print("Calinski-Harabasz:", ch)
print("Davies-Bouldin:", db)

# PCA Visualization
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

plt.figure(figsize=(8,6))
plt.scatter(X_pca[:,0], X_pca[:,1], c=agg_labels, s=50, cmap='plasma')
plt.title("Agglomerative Clusters (PCA Visualization)")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Evaluasi masing-masing model
models = ['K-Means', 'GMM', 'Agglomerative']

sil_scores = [
    silhouette_score(X_scaled, kmeans_labels),
    silhouette_score(X_scaled, gmm_labels),
    silhouette_score(X_scaled, agg_labels)
]

ch_scores = [
    calinski_harabasz_score(X_scaled, kmeans_labels),
    calinski_harabasz_score(X_scaled, gmm_labels),
    calinski_harabasz_score(X_scaled, agg_labels)
]

db_scores = [
    davies_bouldin_score(X_scaled, kmeans_labels),
    davies_bouldin_score(X_scaled, gmm_labels),
    davies_bouldin_score(X_scaled, agg_labels)
]

# Buat bar chart
x = np.arange(len(models))
width = 0.25

fig, ax = plt.subplots(figsize=(10,6))

bar1 = ax.bar(x - width, sil_scores, width, label='Silhouette Score', color='skyblue')
bar2 = ax.bar(x, ch_scores, width, label='Calinski–Harabasz Index', color='lightgreen')
bar3 = ax.bar(x + width, db_scores, width, label='Davies–Bouldin Index', color='salmon')

# Label dan title
ax.set_xlabel('Clustering Model')
ax.set_ylabel('Score')
ax.set_title('Perbandingan Metrik Evaluasi Clustering')
ax.set_xticks(x)
ax.set_xticklabels(models)
ax.legend()

# Tambahkan nilai di atas bar
def add_values(bars):
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f'{height:.3f}',
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0,3),
                    textcoords='offset points',
                    ha='center', va='bottom')

add_values(bar1)
add_values(bar2)
add_values(bar3)

plt.show()
```
